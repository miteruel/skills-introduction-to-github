```python



def extract_text_and_info(pdf_path):
    """
    Función para extraer texto y información general del PDF.        
    """
    try:
        with pdfplumber.open(pdf_path) as pdf:
            text_content = ""
            # Extraer texto de cada p�gina
            for page in pdf.pages:
                text = page.extract_text()
                # Aquí puedes procesar el texto según tus necesidades
                # print(f"Texto de la página: {text}")
            
                text_content += text + "\n"
            return text_content                
    except Exception as e:
        return {"error": f"Error al procesar el PDF: {str(e)}"}


def run1(pdf_path):
    runSimple(pdf_path,extract_text_and_info)
    
run1(ejemplo)
```
