````python
# cogito:14b


import pdfplumber
from tabulate import tabulate

class PDFProcessor:
    def __init__(self, file_path):
        self.file_path = file_path
        try:
            with pdfplumber.open(self.file_path) as pdf:
                self.pages = pdf.pages
                self.number_of_pages = len(self.pages)
        except Exception as e:
            print(f"Error al abrir el archivo: {e}")
            raise

    def extract_basic_info_and_text(self):
        """
        Función para extraer información básica y texto del PDF
        Retorna un diccionario con la estructura del documento
        """
        try:
            pdf_structure = {
                'metadata': {},
                'text_content': [],
                'number_of_pages': self.number_of_pages,
                'errors': []
            }
            
            with pdfplumber.open(self.file_path) as pdf:
                for i, page in enumerate(pdf.pages):
                    # Extraer metadatos de la página
                    metadata = page.get_text("metadata")
                    if metadata:
                        pdf_structure['metadata'][f"page_{i+1}"] = metadata
                    
                    # Extraer texto
                    text = page.extract_text()
                    if text:
                        pdf_structure['text_content'].append({
                            'page': i + 1,
                            'content': text
                        })
                    else:
                        pdf_structure['errors'].append(f"Página {i+1}: No se pudo extraer texto")
            
            return pdf_structure
        
        except Exception as e:
            print(f"Error al procesar el PDF: {e}")
            raise

    def extract_tables_as_json(self):
        """
        Función para extraer tablas del PDF en formato JSON
        Retorna una lista de diccionarios con la estructura de las tablas
        """
        try:
            tables_data = []
            
            with pdfplumber.open(self.file_path) as pdf:
                for i, page in enumerate(pdf.pages):
                    # Extraer todas las tablas de la página actual
                    tables = page.extract_tables()
                    
                    if tables:
                        for table_index, table in enumerate(tables):
                            table_data = {
                                'page_number': i + 1,
                                'table_number': table_index + 1,
                                'data': []
                            }
                            
                            # Convertir cada fila a un diccionario
                            headers = [header.strip() if header else None for header in table[0]]
                            for row in table[1:]:
                                data_row = {}
                                for col, cell_value in enumerate(row):
                                    if headers[col]:
                                        data_row[headers[col]] = cell_value.strip()
                                if data_row:
                                    table_data['data'].append(data_row)
                            
                            tables_data.append(table_data)
            
            return {'tables': tables_data}
        
        except Exception as e:
            print(f"Error al extraer tablas: {e}")
            raise

# Función de ejemplo para mostrar el uso
def main():
    try:
        # Ejemplo de uso
        pdf_processor = PDFProcessor("poliza.pdf")
        
        # Usar la función 1
        basic_info = pdf_processor.extract_basic_info_and_text()
        print("\nInformación básica y texto extraído:")
        print(tabulate([['Páginas', str(basic_info['number_of_pages'])]], 
                      headers=['Tipo', 'Valor']))
        print("\nTexto de las páginas (primeras líneas):")
        for page in basic_info['text_content'][:3]:  # Mostrar solo las primeras 3 páginas
            print(f"Página {page['page']}:")
            print(page['content'][:100] + "...\n")  # Muestra las primeras 100 caracteres
            
        # Usar la función 2
        tables = pdf_processor.extract_tables_as_json()
        if tables:
            print("\nTablas extraídas (formato JSON):")
            for table in tables['tables'][:3]:  # Mostrar solo las primeras 3 tablas
                print(f"Tabla en página {table['page_number']}, tabla {table['table_number']}")
                print("Datos:")
                print(tabulate(table['data'], headers='keys', tablefmt='fancy_grid'))
                
    except Exception as e:
        print(f"Error al ejecutar el programa: {e}")

if __name__ == "__main__":
    main()
    
cogito_14b_help ='''

Este código implementa:

1. Una clase `PDFProcessor` que encapsula las dos funciones principales.

2. La primera función (`extract_basic_info_and_text`) extrae información básica como metadatos y texto, organizándolo en un diccionario estructurado.

3. La segunda función (`extract_tables_as_json`) extrae todas las tablas del PDF y las convierte en una lista de diccionarios con formato JSON, incluyendo la numeración de página y tabla.

4. Incluye manejo de errores para cada sección crítica.

5. Utiliza `tabulate` para mostrar los resultados de forma más legible (necesitarás instalarlo con pip: `pip install tabulate`).

Para usar el programa:
1. Instala las dependencias necesarias:
   ```bash
   pip install pdfplumber tabulate
   ```

2. Ajusta la ruta del archivo PDF en la llamada a la clase.

3. Ejecuta el script para ver los resultados.



'''
````

````python

# qwen3:14b
# pip install pdfplumber

import pdfplumber
import json
import os

def extract_basic_info(pdf_path):
    """
    Extrae información básica de un PDF: texto total y cantidad de páginas.
    
    Args:
        pdf_path (str): Ruta del archivo PDF.

    
    Returns:
        dict: Diccionario con 'text' (texto total) y 'num_pages' (número de páginas).
    """
    if not os.path.exists(pdf_path):
        raise FileNotFoundError(f"El archivo {pdf_path} no existe.")
    
    text = ""
    num_pages = 0
    
    with pdfplumber.open(pdf_path) as pdf:
        num_pages = len(pdf.pages)
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    
    return {
        "text": text.strip(),
        "num_pages": num_pages
    }

def extract_tables_to_json(pdf_path):
    """
    Extrae todas las tablas de un PDF y las devuelve en formato JSON.
    
    Args:
        pdf_path (str): Ruta del archivo PDF.
    
    Returns:
        list: Lista de tablas en formato JSON. Cada tabla es una lista de diccionarios.
    """
    if not os.path.exists(pdf_path):
        raise FileNotFoundError(f"El archivo {pdf_path} no existe.")
    
    tables_json = []
    
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            tables = page.extract_tables()
            for table in tables:
                # Convertir la tabla en una lista de diccionarios con claves numéricas
                headers = [f"cell_{i}" for i in range(len(table[0]))]
                table_json = []
                for row in table:
                    table_json.append(dict(zip(headers, row)))
                tables_json.append(table_json)
    
    return tables_json

# 📌 Ejemplo de uso
if __name__ == "__main__":
    pdf_file = "poliza.pdf"  # Reemplaza con la ruta de tu archivo PDF
    
    try:
        # Ejecutar la extracción básica
        basic_info = extract_basic_info(pdf_file)
        print("=== Información Básica ===")
        print(f"Texto extraído (primeros 200 caracteres): {basic_info['text'][:200]}...")
        print(f"Número de páginas: {basic_info['num_pages']}\n")
        
        # Ejecutar la extracción de tablas
        tables = extract_tables_to_json(pdf_file)
        print("=== Tablas en JSON ===")
        for i, table in enumerate(tables, 1):
            print(f"\nTabla {i}:")
            print(json.dumps(table, indent=2, ensure_ascii=False))
    
    except Exception as e:
        print(f"Error: {e}")

qwen3_14b_help = '''

### 📊 Explicación del Código

1. **`extract_basic_info(pdf_path)`**:
   - Abre el PDF con `pdfplumber`.
   - Extrae el texto de cada página.
   - Devuelve el texto total y el número de páginas.

2. **`extract_tables_to_json(pdf_path)`**:
   - Extrae todas las tablas de cada página.
   - Convierte cada tabla en una lista de diccionarios con claves como `cell_0`, `cell_1`, etc.
   - Devuelve las tablas en formato JSON.

---

### ✅ Ejemplo de Salida (simulado)

```text
=== Información Básica ===
Texto extraído (primeros 200 caracteres): Este documento contiene información sobre...
Número de páginas: 5

=== Tablas en JSON ===

Tabla 1:
[
  {
    "cell_0": "Nombre",
    "cell_1": "Edad",
    "cell_2": "Ciudad"
  },
  {
    "cell_0": "Juan",
    "cell_1": "30",
    "cell_2": "Madrid"
  }
]
```

---

### 🔍 Notas
- **Limitaciones**: `pdfplumber` no extrae texto de imágenes o gráficos. Para PDFs complejos, considera usar `PyMuPDF` o `camelot-py`.
- **Estructura de tablas**: Si las tablas tienen encabezados definidos (como "Nombre", "Edad"), se puede personalizar la clave de los diccionarios.

'''
````
